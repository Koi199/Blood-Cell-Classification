Script to count number of classes

Imports

```python
import pandas as pd
import numpy as np

df = pd.read_csv('LabelledData/LabelExport_20260302.csv')

classes_name = ['Monocyte without RBC', 'Monocyte with RBC', 'RBC alone', 'Lymphocyte?', 'Unusable', 'Clustered cell']

for name in classes_name:
    count = (df['choice'] == name).sum()
    print(f"{name} count: {count}")

print(f"Number of labelled cells: {len(df)-1}")
```
